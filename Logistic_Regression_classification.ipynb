```python
# importing some data science modules

import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Defining the dataset that i took from kaggle 
dataset = pd.read_csv('heart-disease.csv')
dataset
```

```python
# The dependent variable 
x = dataset.drop('target', axis=1)
```

```python
# The indepenent variable 
y = dataset['target']
```

```python
x
```

```python
y
```

```python
# Spliting the dataset into training and test set
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
# We are going to use LogisticRegression for classification from sklearn

from sklearn.linear_model import LogisticRegression
clf = LogisticRegression()
clf
```

```python
# Feeding the data into the algorithm
clf.fit(x_train, y_train)
```

```python
x_test
```

```python
y_test
```

```python
# Predicting the value from the test set
# and it's correct
clf.predict([[60,1, 0, 117, 230, 1, 1, 160, 1, 1.4, 2, 2,3]])
```

```python
# Testing the algorithm on the test_set 
# it is going to output the accuracy, 1 is the max you can get means 100%
clf.score(x_test, y_test)
```
